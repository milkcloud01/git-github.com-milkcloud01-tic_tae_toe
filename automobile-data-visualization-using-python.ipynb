```python
import pandas as pd 
import seaborn as sns 
sns.set(color_codes=True)
```

```python
auto = pd.read_csv('../input/automobile-dataset/Automobile_data.csv')
```

```python
auto.head()
```

```python
sns.distplot(auto['engine-size'])
```

```python
sns.distplot(auto['city-mpg'])
```

```python
sns.jointplot(auto['engine-size'],auto['horsepower'])
```
